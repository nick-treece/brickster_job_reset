```r
install.packages("devtools", dependencies=T)
devtools::install_github("databrickslabs/brickster@issue-131-jobs")
library(brickster)
```

```r
job_name <- "Test job"

notebook_task_1 <- list(
  notebook_path = "Task1",
  source = "GIT"
)

notebook_task_2 <- list(
      notebook_path = "Task2",
      source = "GIT"
)

task1 <- list(
  task_key = "task1",
  existing_cluster_id="0430-105349-vgniypl0",
  notebook_task = notebook_task_1
)

task2 <- list(
    task_key = "Task2",
    existing_cluster_id = "0430-105349-vgniypl0",
    notebook_task = notebook_task_2
)

parameters <- list(
  parameter1 = "Test"
)

tasks <- list()
tasks <- append(tasks, list(task1))
tasks <- append(tasks, list(task2))

git_source = list(
                  git_url = "https://github.com/nick-treece/brickster_job_reset.git",
                  git_provider = "gitHub",
                  git_branch = "main"
                )
```

```r
response <- db_jobs_create(name = job_name,
              tasks = tasks,
              parameters = parameters,
              git_source = git_source
)
response
```

```r
response <- db_jobs_reset(job_id = response$job_id,
                          name = job_name,
                          tasks = tasks,
                          parameters = parameters,
                          git_source = git_source
)
```
